# 0. Expected Behavior

* **(Reduction:)** optimal value, feasible, and usually the slowest (most sensitive to problem structure).
* **(Approximation:)** â‰¥ optimal value, infeasible but $cost \leq (1+\epsilon)*B$, and very fast. Gets faster with *larger* $\epsilon$.
* **(Feasible Heuristic:)** Feasible, with speeds slightly faster than Approximation of same $\epsilon$. Not optimal, but value is often $\geq \frac{1}{1+\epsilon} OPT$.
* **(RL Agent:)** Very fast, always feasible, value is unpredictable. May be optimal on some distributions and do poorly on others.
