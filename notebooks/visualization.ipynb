# 0. Expectations

### We expect the following behavior:
* **\[Reduction]** Should be exactly optimal and never violate the budget. However, it should usually be the slowest method. Note, since it is so slow, we run it for smaller H values than the other solvers just to show how much better the approximations are.
* **\[Approximation]** Should be at least optimal value, but violate budget by a bit (no more than (1+epsilon)*budget). It should become faster, for larger choices of epsilon (1 being the fastest).
* **\[Feasible Heuristic]** Should always be feasible, but may not be optimal. For most instance it should achieve at least half optimal value or better (as epsilon gets smaller). The speed should be nearly identical to Approximation under the same epsilon.
